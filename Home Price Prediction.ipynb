### The following steps cluster the neighborhood into three groups: 
>1. low price; 
>2. high price low frequency; 
>3. high price high frequency.

```python
# Calculate frequency counts of neighborhoods
freq = data['neighborhood'].value_counts().reset_index()
freq.columns = ['neighborhood', 'freq']

# Calculate the mean price per square foot for each neighborhood
mean = data.groupby('neighborhood')['price_per_sqft'].mean().reset_index()

# Merge the frequency counts and mean prices
cluster = pd.merge(freq, mean, on='neighborhood')
cluster.describe()

# Plot frequency counts
plt.figure(figsize=(12, 6))
plt.bar(cluster['neighborhood'], cluster['freq'])
plt.xlabel('Neighborhood')
plt.ylabel('Frequency Count')
plt.title('Frequency Counts of Neighborhoods')
plt.xticks(rotation=90)
plt.show()

# Plot mean price per square foot
plt.figure(figsize=(12, 6))
plt.bar(cluster['neighborhood'], cluster['price_per_sqft'])
plt.xlabel('Neighborhood')
plt.ylabel('Mean Price per Sqft')
plt.title('Mean Price per Sqft by Neighborhood')
plt.xticks(rotation=90)
plt.show()
```

### These are the low price neighborhoods:

```python
cluster1=cluster[cluster.price_per_sqft<756]
cluster1.index

# Create a bar chart to visualize the frequency counts
plt.figure(figsize=(10, 6))
plt.bar(cluster1['neighborhood'], cluster1['freq'])
plt.xlabel('Neighborhood')
plt.ylabel('Frequency')
plt.title('Frequency Counts of Neighborhoods in cluster1')
plt.xticks(rotation=90)
plt.show()
```

### These are the high price and low frequency neighborhoods:

```python
cluster_temp=cluster[cluster.price_per_sqft>=756]
cluster2=cluster_temp[cluster_temp.freq<123]
cluster2.index

plt.figure(figsize=(10, 6))
plt.bar(cluster2['neighborhood'], cluster2['freq'])
plt.xlabel('Neighborhood')
plt.ylabel('Frequency')
plt.title('Frequency Counts of Neighborhoods in cluster2')
plt.xticks(rotation=90)
plt.show()
```

### These are the high price and high frequency neighborhoods:

```python
cluster3=cluster_temp[cluster_temp.freq>=123]
cluster3.index

plt.figure(figsize=(10, 6))
plt.bar(cluster3['neighborhood'], cluster3['freq'])
plt.xlabel('Neighborhood')
plt.ylabel('Frequency')
plt.title('Frequency Counts of Neighborhoods in cluster3')
plt.xticks(rotation=90)
plt.show()
```

### We add a group column based on the clusters:



```python
def get_group(x):
    if x in cluster1.index:
        return 'low_price'
    elif x in cluster2.index:
        return 'high_price_low_freq'
    else:
        return 'high_price_high_freq'

data['group']=data.neighborhood.apply(get_group)
```

```python
data.drop(data.columns[[0,4,6,7,8,13]],axis=1,inplace=True)
data=data[['bathrooms','bedrooms','finishedsqft','totalrooms','usecode','yearbuilt','zindexvalue','group','lastsoldprice']]
data.head()
```

```python
X=data[['bathrooms','bedrooms','finishedsqft','totalrooms','usecode','yearbuilt','zindexvalue','group']]
Y=data['lastsoldprice']

#dummy variable
n=pd.get_dummies(data.group)
X=pd.concat([X,n],axis=1)
m=pd.get_dummies(data.usecode)
X=pd.concat([X,m],axis=1)
drops=['group','usecode']
X.drop(drops,inplace=True,axis=1)
X.head()
```

# Train and Build Model

# 1. Linear Regression model

### Split the data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

### Train the model

```python
from sklearn.linear_model import LinearRegression

reg=LinearRegression()
reg.fit(X_train,y_train)
```

## Calculate R squared:



```python
y_pred=reg.predict(X_test)
print('Linear Regression R squared: %.4f'%reg.score(X_test,y_test))
```

## Calculate root-mean-square error (RMSE):



```python
from sklearn.metrics import mean_squared_error

#find mean square error
lin_mse=mean_squared_error(y_pred,y_test)

# find root mean square error
lin_rmse=np.sqrt(lin_mse)

print('Linear Regression RMSE: %.4f'%lin_rmse)

#checking for linearity
# Create a scatter plot of actual vs. predicted values
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_test, c='blue', edgecolors='k')
plt.scatter(y_test, y_pred, c='red', edgecolors='k')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs. Predicted Values (Linear Regression)')
plt.grid(True)
plt.show()
```

## Calculate mean absolute error (MAE):



```python
from sklearn.metrics import mean_absolute_error

lin_mae=mean_absolute_error(y_pred,y_test)
print("Linear Regression MAE: %.4f"%lin_mae)
```

# 2. Random Forest Model

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg=RandomForestRegressor(random_state=42)
#fit the model
forest_reg.fit(X_train,y_train)
```

```python
print("Random Forest R Sqaured: %.4f"%forest_reg.score(X_test,y_test))
```

```python
y_pred=forest_reg.predict(X_test)
# Mean Squared Error
forest_mse=mean_squared_error(y_pred,y_test)
print("Random Forest MSE  : %.4f"%forest_mse)

# root mean sqaured error
forest_rmse=np.sqrt(forest_mse)
print("Random Forest RMSE : %.4f"%forest_rmse)

#checking for linearity
# Create a scatter plot of actual vs. predicted values
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_test, c='blue', edgecolors='k')
plt.scatter(y_test, y_pred, c='red', edgecolors='k')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs. Predicted Values (Random Forest)')
plt.grid(True)
plt.show()
```

# 3. Gradient Boosting Model

```python
from sklearn.ensemble import GradientBoostingRegressor

model=GradientBoostingRegressor()
# fit train data into model
model.fit(X_train,y_train)
```

### R sqaured

```python
print("Gradient Boosting R sqaured: %.4f"%model.score(X_test,y_test))
```

### MSE & RMSE errors

```python
y_pred=model.predict(X_test)

# MSE
model_mse=mean_squared_error(y_pred,y_test)
print("Gradient Boosting MSE : %.4f"%model_mse)

#RMSE
model_rmse=np.sqrt(model_mse)
print("Gradient Boosting RMSE: %.4f"%model_rmse)

# Create a scatter plot of actual vs. predicted values
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_test, c='blue', edgecolors='k')
plt.scatter(y_test, y_pred, c='red', edgecolors='k')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs. Predicted Values (Gradient Boosting)')
plt.grid(True)
plt.show()
```
